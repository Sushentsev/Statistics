## Равномерное распределение

Показали, что оценка параметра $\theta$ из $U[0, \theta]$ с помощью $k$-го момента имеет вид: $\theta^* = ((k + 1) \cdot \overline{X^k})^\frac{1}{k}$.

Параметры: $sample\_size = 1000, samples\_count = 100, moments = [1; 1000], \theta = 1$.

```python
sample_size = 1000
samples_count = 100
moments = range(1, 1000 + 1)
theta = 1
```

```python
plot_unif(sample_size, samples_count, moments, theta)
```

Построим логарифм ошибки, чтобы лучше рассмотреть.

```python
plot_unif_log(sample_size, samples_count, moments, theta)
```

Можно заметить, что величина ошибки убывает с ростом $k$, поэтому выгоднее брать $k$ как можно больше.

## Показательное распределение

Показали, что оценка параметра $\theta$ из $Exp(\theta)$ с помощью $k$-го момента имеет вид: $\theta^* = \left(\frac{\overline{X^k}}{k!}\right)^\frac{1}{k}$.

Параметры: $sample\_size = 1000, samples\_count = 100, moments = [1; 100], \theta = 1$.

```python
sample_size = 1000
samples_count = 100
moments = range(1, 100 + 1)
theta = 1
```

```python
plot_exp(sample_size, samples_count, moments, theta)
```

Можно заметить, что величина ошибки возрастает с ростом $k$, поэтому выгоднее брать $k$ как можно меньше.
